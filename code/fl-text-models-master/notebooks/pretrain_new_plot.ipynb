# Part 1: Small Network Loss and Accuracy

### Load Accuracy and Loss Arrays from Small Model (Batch Size 16)

```python
network = '100_256_100'
```

```python
random_train_accuracy = np.load('embedding_layer_results/random_{}/train_accuracy.npy'.format(network))[0:500]
random_val_accuracy = np.load('embedding_layer_results/random_{}/val_accuracy.npy'.format(network))[0:500]
```

```python
so_finetune_train_accuracy = np.load('embedding_layer_results/train_accuracy_finetuning.npy')
so_finetune_val_accuracy = np.load('embedding_layer_results/val_accuracy_finetuning.npy')
```

```python
sp_finetune_train_accuracy = np.load('embedding_layer_results/train_accuracy_finetuning_sp.npy')
sp_finetune_val_accuracy = np.load('embedding_layer_results/val_accuracy_finetuning_sp.npy')
```

```python
baseline_train_accuracy = np.load('embedding_layer_results/train_accuracy_no_pt_embed.npy')
baseline_val_accuracy = np.load('embedding_layer_results/val_accuracy_no_pt_embed.npy')
```

```python
x_axis = range(0, len(random_train_accuracy))
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.plot(x_axis, baseline_train_accuracy, label='Train with No Pretraining', linestyle=':')
ax.plot(x_axis, baseline_val_accuracy, label='Validation with No Pretraining', linestyle=':')
ax.plot(x_axis, so_finetune_train_accuracy, label='Train with Stack Overflow Pretraining', linestyle='-.')
ax.plot(x_axis, so_finetune_val_accuracy, label='Validation with Stack Overflow Pretraining', linestyle='-.')
ax.plot(x_axis, sp_finetune_train_accuracy, label='Train with Shakespeare Pretraining', linestyle='--')
ax.plot(x_axis, sp_finetune_val_accuracy, label='Validation with Shakespeare Pretraining', linestyle='--')
ax.legend(loc='best', prop={'size': 15})
plt.title('Small Network Full Model Pretraining Experiments', fontsize=20)
plt.xlabel('Epochs', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
plt.show()
```
